```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the CSV file containing execution times
df = pd.read_csv("sorting_times.csv")

# Create a plot for each sorting algorithm
plt.figure(figsize=(10, 6))

# Insertion Sort
plt.plot(df["ArraySize"], df["InsertionSortTime"], label="Insertion Sort")

# Merge Sort (Fusion Sort)
plt.plot(df["ArraySize"], df["FusionSortTime"], label="Merge Sort")

# Java's Arrays.sort()
plt.plot(df["ArraySize"], df["JavaSortTime"], label="Java Arrays.sort()")

plt.xlabel("Array Size")
plt.ylabel("Execution Time (milliseconds)")
plt.title("Sorting Algorithm Comparison (Logarithmic Scale)")
plt.yscale("log")  # Set the y-axis to a logarithmic scale
plt.legend()
plt.grid(True)

plt.savefig("sorting_comparison_log.png")
plt.show()
```
